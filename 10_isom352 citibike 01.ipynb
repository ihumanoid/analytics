```python
query = "SELECT rideable_type, "\
        "avg(cast(ended_at AS datetime)-cast(started_at AS datetime)) as rentdur "\
        "FROM citibike.tripdata202209 "\
        "GROUP BY rideable_type;"
cursor.execute(query)
df = pd.DataFrame(cursor.fetchall())
df.columns=['rideable_type', 'rentdur_avg']
df
```

```python
fig = sns.scatterplot(data=df, x='rideable_type', y='rentdur_avg')
```

```python
query = "SELECT start_station_name, "\
        "count(*) as numrentals "\
        "FROM citibike.tripdata202209 "\
        "GROUP BY start_station_name "\
        "ORDER BY numrentals DESC LIMIT 10;"
cursor.execute(query)
df = pd.DataFrame(cursor.fetchall())
df.columns=['start_station_name', 'numrentals']
df
```
