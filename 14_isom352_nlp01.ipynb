```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import STOPWORDS

#read data
df = pd.read_csv('datasets/Women Dresses Reviews (clean).csv')

#combine column data into one big text
df2 = df[df['review_text'].str.contains("bad")]['review_text']
text = " ".join(df2)
df2
```

```python
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
import nltk
speech = "If you choose to use your status and influence to raise your voice on behalf of those who have no voice; if you choose to identify not only with the powerful, but with the powerless; if you retain the ability to imagine yourself into the lives of those who do not have your advantages, then it will not only be your proud families who celebrate your existence, but thousands and millions of people whose reality you have helped change. We do not need magic to change the world, we carry all the power we need inside ourselves already: we have the power to imagine better."
tokens = speech.split()
tokens
```

```python
tokens.count("if")
```

```python
from nltk.tokenize import WordPunctTokenizer
tk = WordPunctTokenizer() 
print(tk.tokenize("a..bccd#!efh"))
```

```python
tokens = tk.tokenize(speech.lower())
tokens
```

```python
pos = nltk.pos_tag(tokens)
pos
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
wordnet_lemmatizer.lemmatize("computers")
```

```python
from nltk.util import bigrams
from nltk.util import ngrams
from nltk.util import everygrams

print(list(bigrams(tokens)))
```

```python
print(list(ngrams(tokens, n=3)))
```

```python
print(list(everygrams(tokens, max_len=3)))
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(tokens)
fdist
```

```python
from wordcloud import WordCloud
filter_words = dict([(m, n) for m, n in fdist.items() if len(m) > 4])
wc = WordCloud().generate_from_frequencies(fdist)

plt.imshow(wc)
plt.show()
```
