```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('datasets/gapminder.tsv', sep='\t')
dfmean = df.groupby('country')[['gdpPercap', 'pop', 'lifeExp']].mean().reset_index()
dfmean
```

```python
dfmean2 = dfmean.sort_values('gdpPercap')
plt.plot(dfmean2['country'].tail(), dfmean2['gdpPercap'].tail())
plt.plot(dfmean2['country'].tail(), dfmean2['gdpPercap'].tail(), 'x')
plt.plot(dfmean2['country'].tail(), dfmean2['gdpPercap'].tail(), 'o')
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['tippct']=df['tip']/df['total_bill']
fig = sns.violinplot(data=df, x='day', y='tippct')
```

```python
fig = sns.relplot(data=df, x='total_bill', y='tippct')
fig.set_axis_labels('Total Bill', 'Tip Perecentage')
```

```python
fig = sns.relplot(data=df, x='total_bill', y='tippct', hue='smoker')
fig.set_axis_labels('Total Bill', 'Tip Perecentage')
fig.ax.axline(xy1=(10,.1), xy2=(40,0.2), color='green')
fig.ax.axline(xy1=(10,.1), slope=.5, color='red')
```
